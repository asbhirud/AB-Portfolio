```python
df["Age Group"] = pd.cut(df["Age"], bins=5, labels=["26-30","31-35", "36-40","41-45","45-More"])
```

```python
df.head()
```

```python
pd.crosstab(columns=df["Subscription Type"], index=df["Age Group"])
```

```python
sns.heatmap(pd.crosstab(columns=df["Subscription Type"], index=df["Age Group"]), annot=True, fmt="d", cmap="viridis")
plt.show()
```

# Insights
* By using this we can clearly see that the basic subscription is more often subscribed by 26-30 age group
* In 41=45 age group the subscription is less compared to other and that is Premium type
